```r
#PTHA18 library
library(ncdf4)
library(rptha)
```

```r
ptha18 = new.env()
source('../ptha-2/ptha_access/get_PTHA_results.R', local=ptha18, chdir=TRUE)
```

```r
# Get the full PTHA18 source-zone events table, unit-source data, etc
source_zone = 'sunda2'
sz_HS_scenarios = ptha18$get_source_zone_events_data(source_zone, slip_type='stochastic')

# Only get the unsegmented random scenarios
sampled_scenarios = read.csv("random_scenarios_sunda2_andaman_segment_HS.csv")
all_scenarios = sampled_scenarios  # Just the unsegmented scenarios
```

```r


# Skip the NCI-specific copying becuse using local path

cache_dir = paste0('BARU2/SOURCE_ZONES/', source_zone, '/EQ_SOURCE/Unit_source_data/', source_zone)
dir.create(cache_dir, recursive=TRUE, showWarnings=FALSE)

# Store the initial conditions here
output_tif_dir = 'BARU2/scenario_initial_conditions_andaman'
dir.create(output_tif_dir, showWarnings=FALSE)

# Create the raster for the ith unique_scenario_id
make_initial_condition_i <- function(i) {
    target_row = unique_scenario_ids[i]
    expected_Mw = unique_scenario_Mws[i]
    local_Mw = sz_HS_scenarios$events$Mw[target_row]
    if(round(local_Mw, 1) != round(expected_Mw, 1)) stop('Logic error -- these magnitudes should be identical')
    
    # Make a row index character, padded with leading zeros as required, for consistency
    # of file names
    target_row_fixed_width = substring(as.character(1e+07 + target_row), 2, 8)
    output_filename = paste0(output_tif_dir, '/', source_zone, '_row_', 
                             target_row_fixed_width, 
                             '_Mw_', round(local_Mw*10), '_HS.tif')
    
    # Only generate if it doesn't already exist
    if(!file.exists(output_filename)) {
        message(paste("Generating initial condition for scenario", i, "of", length(unique_scenario_ids)))
        r1 = ptha18$get_initial_condition_for_event(sz_HS_scenarios, event_ID=target_row)
        writeRaster(r1, output_filename, options=c('COMPRESS=DEFLATE'), overwrite=TRUE)
        rm(r1); gc()
    } else {
        message(paste("File already exists:", output_filename))
    }
}

# Get unique scenario IDs from the unsegmented data
unique_scenario_ids = unique(all_scenarios$inds)
unique_scenario_Mws = all_scenarios$mw[match(unique_scenario_ids, all_scenarios$inds)]

# Try one first to make sure it works
if(length(unique_scenario_ids) > 0) {
    message("Trying to generate one initial condition as a test...")
    make_initial_condition_i(1)
    message("First initial condition generated successfully. Proceeding with the rest.")
    
    # Only continue with parallel processing if the first one worked
    library(parallel)
    MC_CORES = min(30, parallel::detectCores())  # Adjust based on available cores
    mclapply(2:length(unique_scenario_ids), make_initial_condition_i, mc.cores=MC_CORES)
} else {
    message("No unique scenario IDs found in the unsegmented dataset.")
}
```
